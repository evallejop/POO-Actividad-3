```python
import tkinter as tk
from tkinter import messagebox
import statistics


class Notas:
    def __init__(self):
        self.lista_notas = []

    def calcular_promedio(self):
        return statistics.mean(self.lista_notas)

    def calcular_desviacion(self):
        return statistics.stdev(self.lista_notas)

    def calcular_mayor(self):
        return max(self.lista_notas)

    def calcular_menor(self):
        return min(self.lista_notas)


class Ventanaprincipal:
    def __init__(self, ventana):
        self.ventana = ventana
        self.ventana.title("Calculadora de Notas")
        self.ventana.geometry("300x450")
        self.ventana.resizable(False, False)
      
        self.modelo_notas = Notas()
        
        self.entrada_notas = []
        self.nombres_notas = ["Nota 1", "Nota 2", "Nota 3", "Nota 4", "Nota 5"]
        self.inicio()

    def inicio(self):
        for i, nombre in enumerate(self.nombres_notas):
            tk.Label(self.ventana, text=f"{nombre}:").grid(row=i, column=0, padx=10, pady=5, sticky="w")
            entrada = tk.Entry(self.ventana, width=15)
            entrada.grid(row=i, column=1, padx=10, pady=5)
            self.entrada_notas.append(entrada)

        # ---Botones---
        self.boton_calcular = tk.Button(self.ventana, text="Calcular", command=self.action_performed_calcular, 
                                        bg="#4CAF50", fg="white", font=("Arial", 10, "bold"))
        self.boton_calcular.grid(row=len(self.nombres_notas), column=0, pady=15, padx=10, sticky="ew")

        self.boton_limpiar = tk.Button(self.ventana, text="Limpiar", command=self.limpiar_campos, 
                                       bg="#f44336", fg="white", font=("Arial", 10, "bold"))
        self.boton_limpiar.grid(row=len(self.nombres_notas), column=1, pady=15, padx=10, sticky="ew")

        # ---Etiquetas---
        self.resultado_promedio = tk.Label(self.ventana, text="Promedio:", font=("Arial", 10))
        self.resultado_promedio.grid(row=6, column=0, columnspan=2, pady=5, sticky="w")

        self.resultado_desviacion = tk.Label(self.ventana, text="Desviación Estándar:", font=("Arial", 10))
        self.resultado_desviacion.grid(row=7, column=0, columnspan=2, pady=5, sticky="w")

        self.resultado_mayor = tk.Label(self.ventana, text="Nota Mayor:", font=("Arial", 10))
        self.resultado_mayor.grid(row=8, column=0, columnspan=2, pady=5, sticky="w")

        self.resultado_menor = tk.Label(self.ventana, text="Nota Menor:", font=("Arial", 10))
        self.resultado_menor.grid(row=9, column=0, columnspan=2, pady=5, sticky="w")

    def action_performed_calcular(self):
        notas_temp = []
        try:
            for entry in self.entrada_notas:
                valor = float(entry.get())
                if 0 <= valor <= 5:
                    notas_temp.append(valor)
                else:
                    messagebox.showerror("Error de entrada", "Todas las notas deben estar entre 0 y 5.")
                    return

      
            self.modelo_notas.lista_notas = notas_temp

          
            self.resultado_promedio.config(text=f"Promedio: {self.modelo_notas.calcular_promedio():.2f}")
            self.resultado_desviacion.config(text=f"Desviación Estándar: {self.modelo_notas.calcular_desviacion():.2f}")
            self.resultado_mayor.config(text=f"Nota Mayor: {self.modelo_notas.calcular_mayor():.2f}")
            self.resultado_menor.config(text=f"Nota Menor: {self.modelo_notas.calcular_menor():.2f}")

        except ValueError:
            messagebox.showerror("Error de Entrada", "Asegúrate de ingresar números válidos.")

    def limpiar_campos(self):
        """Gestiona el evento de limpieza"""
        for entry in self.entrada_notas:
            entry.delete(0, tk.END)
        self.resultado_promedio.config(text="Promedio:")
        self.resultado_desviacion.config(text="Desviación Estándar:")
        self.resultado_mayor.config(text="Nota Mayor:")
        self.resultado_menor.config(text="Nota Menor:")


if __name__ == "__main__":
    ventana_tk = tk.Tk()
    app = Ventanaprincipal(ventana_tk)
    ventana_tk.mainloop()
```
